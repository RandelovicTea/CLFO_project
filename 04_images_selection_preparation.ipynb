### Selecting files
- Number of images for each category has to be the same
- Uninverted images are filtered
- All eyes belong to same side 


```python
# with notch, right eye, 5x100 for each class
outputs = pd.read_csv("d:/project_data/trainLabels.csv")

num_classes = 5
img_paths = [ [] for _ in range (num_classes)]

proceed = 5
level_status = [0,0,0,0,0]

dir_path = 'd:\\project_data\\train_1'
image_files=os.listdir(dir_path)

data = selector(image_files, outputs)

row = 0
while proceed:
    try:
        #print(data["level"].value_counts())
        #print(len(image_files)
        level = int(data["level"][row])
        
        if level_status[level] < 100:
            abs_path = os.path.join(dir_path, data['image'][row]) + '.jpeg'
            if notch_selector(abs_path) and 'right' in abs_path:
                img_paths[level].append(abs_path)
                level_status[level]+=1 

        elif level_status[level] == 100:
            proceed-=1
            level_status[level]+=1      
    except:
        pass
    
    if row%10==0:
        print(f'{row}/{100*num_classes}') 
    row+=1
    
print(level_status)
```

```python
import pickle 

with open('notchers_5x100', 'wb') as f:
    pickle.dump(img_paths, f)
```
